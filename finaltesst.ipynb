```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

# Khởi tạo WebDriver
driver = webdriver.Chrome()

# Khởi tạo danh sách để lưu thông tin
painters_info = []

# Lặp qua các trang bắt đầu từ ký tự 'A' (mã ASCII 65) đến 'Z' (mã ASCII 90)
for i in range(65, 91):  # Bạn có thể thay 66 bằng 91 để duyệt từ A-Z
    url = "https://en.wikipedia.org/wiki/List_of_painters_by_name_beginning_with_%22" + chr(i) + "%22"
    driver.get(url)
    time.sleep(10)
    
    # Tìm tất cả các họa sĩ trong danh sách
    painters = driver.find_elements(By.XPATH, '//div[@class="div-col"]//li')

    # Duyệt qua danh sách họa sĩ và lấy thông tin
    for painter in painters:
        text = painter.text
        parts = text.split(',')
        if len(parts) >= 2:
            name_age = parts[0].split('(')
            name = name_age[0].strip()
            if len(name_age) == 1:
                birth_death = None
            else:
                birth_death = name_age[1].strip(")")
                if all(char not in '0123456789' for char in birth_death) and len(parts) >= 3:
                    birth_death = parts[1].strip(")")
                    full_info = parts[2].split(" ")
                    nationality = full_info[1].strip()
                else:
                    full_info = parts[1].split(" ")
                    nationality = full_info[1].strip()
        else:
            nationality = None
            birth_death = None

        # Lấy liên kết của họa sĩ
        try:
            link_element = painter.find_element(By.TAG_NAME, 'a')
            link = link_element.get_attribute('href')
        except:
            continue
        # Thêm thông tin họa sĩ vào danh sách
        painters_info.append({
            'name': name,
            'birth_death': birth_death,
            'nationality': nationality,
            'link': link
        })

# Đóng WebDriver
driver.quit()

# Chuyển danh sách họa sĩ thành DataFrame
df = pd.DataFrame(painters_info)

```
